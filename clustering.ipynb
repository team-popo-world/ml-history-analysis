# MongoDB에서 바로 DataFrame으로 가져오기

```python
from pymongo import MongoClient
import pandas as pd
import os
from dotenv import load_dotenv

load_dotenv(override=True)

uri = os.getenv("MONGO_URI")
db_name = os.getenv("MONGO_DB_NAME")
collection_name = os.getenv("COLLECTION_NAME")

client = MongoClient(uri)
db = client[db_name]
collection = db[collection_name]

# MongoDB에서 데이터 불러오기
data = list(collection.find())

# DataFrame으로 변환
df = pd.DataFrame(data)
```

```python
df.head()
```

# PostgreSQL에서 DataFrame으로 가져오기

```python
import pandas as pd
from dotenv import load_dotenv
import os
import psycopg2

load_dotenv(override=True)

# PostgreSQL 연결 정보
conn = psycopg2.connect(
    host=os.getenv("DB_HOST"),
    port=os.getenv("DB_PORT"),          
    dbname=os.getenv("DB_NAME"), 
    user=os.getenv("DB_USER"),
    password=os.getenv("DB_PASSWORD") 
)

# SQL 쿼리 실행 후 DataFrame으로 불러오기
s_query = "SELECT chapter_id, seed_money FROM invest_chapter;"
# c_query = "SELECT child_id, sex, age FROM child;"
seed = pd.read_sql(s_query, conn)
# child = pd.read_sql(c_query, conn)

# 연결 종료
conn.close()

# 결과 확인
print(seed.head())
# print(child.head())
```

```python
seed.rename(columns={'chapter_id': 'chapterId'}, inplace=True)
seed.rename(columns={'seed_money': 'seedMoney'}, inplace=True)
seed.head()
```

# 거래참여 턴 개수

```python
tpt = df[['investSessionId', 
          'childId',
          'turn',
          'riskLevel',
          'transactionType']].copy()

tpt.head()
```

```python
hold_flags = tpt.groupby(['investSessionId', 'childId', 'turn'])['transactionType'].apply(lambda x: int((x != "KEEP").any())).reset_index(name="notAllKeep")
hold_flags.head()
```

```python
tradingTurn = hold_flags.groupby(['investSessionId', 'childId'])['notAllKeep'].mean().reset_index(name="avgNotKeep")
tradingTurn.head()
```

# 거래횟수

```python
nt = df[['investSessionId', 
          'turn',
          'riskLevel',
          'plusClick',
          'minusClick']].copy()

nt['click'] = nt['plusClick'] + nt['minusClick']

nt.head()
```

```python
transactionNum = nt.groupby('investSessionId')[['click']].mean().reset_index()
transactionNum.head()
```

# 평균 현금잔여율

```python
acr = df[['investSessionId', 
          'chapterId',
          'turn',
          'currentValue']].copy()

acr = acr.groupby(['investSessionId', 'chapterId', 'turn'])['currentValue'].mean().reset_index(name='avgCurrentValue')

acr.head()
acr.info()
```

```python
acr_merged = acr.merge(seed, on='chapterId', how='left')
acr_merged.head()
acr_merged.info()
```

```python
acr_merged['diff'] = acr_merged['seedMoney'] - acr_merged['avgCurrentValue']

averageCashRatio = acr_merged.groupby('investSessionId')['diff'].mean().reset_index(name="avgDiff")

averageCashRatio.head()
```

# 평균 턴 체류시간

```python
# 평균 턴 체류시간
turn = df[['investSessionId', 
                'turn',
                'startedAt',
                'endedAt']].copy()
```

```python
avgStayTime = turn.drop_duplicates(subset=['investSessionId', 'turn']).copy()
```

```python
avgStayTime['stayTime'] = (avgStayTime['endedAt'] - avgStayTime['startedAt']).dt.total_seconds()
avgStayTime.drop(columns=['startedAt', 'endedAt'], inplace=True)

avgStayTime = avgStayTime.groupby('investSessionId')['stayTime'].mean().reset_index(name='avgStayTime')
```

# 전체 병합

```python
merged = tradingTurn.merge(transactionNum, on='investSessionId', how='inner').merge(averageCashRatio, on='investSessionId', how='inner').merge(avgStayTime, on='investSessionId', how='inner')
merged.head()
```
