## 인코딩

### userId 레이블 인코딩

```python
# 레이블 인코딩
from sklearn.preprocessing import LabelEncoder

# LabelEncoder 객체 생성
encoder = LabelEncoder()

# 데이터에 fit_transform 적용하여 인코딩
df['userId'] = encoder.fit_transform(df['userId'])

# 4. 인코딩된 값과 원래 값의 매핑 확인 (선택 사항)
# 이 정보를 저장해두면 나중에 인코딩된 숫자가 무엇을 의미하는지 파악하거나
# 결과를 다시 원본 문자열로 되돌릴 때 유용합니다.
label_mapping = dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))

label_mapping
```

### sex 원핫 인코딩

```python
# 레이블 인코딩
from sklearn.preprocessing import LabelEncoder

# LabelEncoder 객체 생성
encoder = LabelEncoder()

# 데이터에 fit_transform 적용하여 인코딩
df['sex'] = encoder.fit_transform(df['sex'])

# 4. 인코딩된 값과 원래 값의 매핑 확인 (선택 사항)
# 이 정보를 저장해두면 나중에 인코딩된 숫자가 무엇을 의미하는지 파악하거나
# 결과를 다시 원본 문자열로 되돌릴 때 유용합니다.
label_mapping = dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))

label_mapping
```

### chapterId 레이블 인코딩

```python
# 레이블 인코딩
from sklearn.preprocessing import LabelEncoder

# LabelEncoder 객체 생성
encoder = LabelEncoder()

# 데이터에 fit_transform 적용하여 인코딩
df['chapterId'] = encoder.fit_transform(df['chapterId'])

# 4. 인코딩된 값과 원래 값의 매핑 확인 (선택 사항)
label_mapping = dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))

label_mapping
```

### investPro 레이블 인코딩

```python
# 매핑 딕셔너리 정의 (예: "남성"은 0, "여성"은 1로 매핑)
pro_mapping = {'공격투자형': 0, '적극투자형': 1, '위험중립형':2, "안정추구형":3, "안정형":4}

# 'sex' 컬럼에 매핑 적용
df['investPro'] = df['investPro'].map(pro_mapping)

pro_mapping
```

# 데이터 분할

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df.drop(["investSessionId","investPro"], axis=1),
    df['investPro'],
    random_state=42
)
```

# 모델 훈련 및 예측
- 결정트리
- 랜덤 포레스트
- 그래디언트 부스팅
- 로지스틱 회귀 : 원래는 이진인데 OvR / multinomial 전략으로 다중분류 가능
- 서포트 벡터 머신(SVM) : 원래는 이진인데 OvR/OvO 방식으로 다중분류 가능

## 결정트리 모델

```python
## 모델 훈련 ##
from sklearn.tree import DecisionTreeClassifier

# 결정 트리 분류 모델 객체 생성
clf = DecisionTreeClassifier(random_state=42)

# 모델 훈련
clf = clf.fit(X_train, y_train)

# 테스트 데이터로 예측 수행
y_pred = clf.predict(X_test)


## 예측 ##
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# 정확도 계산
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# 정밀도 계산
# TP / (FP + TP) = 양성(Positive)로 예측 했을 때 실제 맞은 비율
precision = precision_score(y_test, y_pred, average='macro')
print("Precision:", precision)

# 재현율 계산
# TP / (FN + TP) = 실제 양성인 데이터를 맞춘 비율
recall = recall_score(y_test, y_pred, average='macro')
print("Recall:", recall)

# f1_score 계산
# 정밀도, 재현율의 조화 평균: 두 지표의 균형을 고려할 때 사용함
f1 = f1_score(y_test, y_pred, average='macro')
print("F1 Score:", f1)
```

```python
# 오차 행렬 시각화 - 히트맵
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# 혼동 행렬 계산
conf_matrix = confusion_matrix(y_test, y_pred)

# 혼동 행렬 시각화
plt.figure(figsize=(12, 5))
plt.subplot(121)
ax = sns.heatmap(
    conf_matrix,
    annot=True,
    cmap="Reds"
)
ax.set_title("DecisionTreeClassifier's confusion matrix")


plt.xlabel("Predict")
plt.ylabel("Actual")
plt.show()
```
